# Data Prepocessing
The data preprocessing steps involve converting Holding Policy duration from categorical to numeric features.

The Region Code is in numeric form, converting it into string and using it as Categorical feature makes more sense.

Creating a feature_region_code by concatenating region_code+city_code.
(This feature was not useful and thus was deleted later)

Converting features like Acommadation type, Is_spouse,Insurance_type to encoded features for further use.

Health indicator can be used as a numeric feature because it follows an ascending order.


engineering a feature upper_age-lower_age-- which gives us age difference between two holders.

this gives us an insight as less difference might mean policy holder are couples or siblings
and large difference means children and parents.

```python
import pandas as pd
df=pd.read_csv('train_Df64byy.csv')

def getNumber(x):
    return str(x)
df["Region_Code"]=df["Region_Code"].apply(getNumber)




#x1 = float("nan")
def getNumber(x):
    if x=="14+":
        return 15
    elif x!=x:
        return 0
    else:
        return int(float(x))
df["Holding_Policy_Duration"]=df["Holding_Policy_Duration"].apply(getNumber)


df['City_code_region']=df['City_Code']+df['Region_Code']


def getNumber(x):
    if x=="Rented":
        return 0
    elif x!=x:
        return 0
    else:
        return 1
df["Accomodation_Type"]=df["Accomodation_Type"].apply(getNumber)



#x1 = float("nan")
def getNumber(x):
    if x=="No":
        return 0
    elif x!=x:
        return 0
    else:
        return 1
df["Is_Spouse"]=df["Is_Spouse"].apply(getNumber)
#We have created a new column called "gender" and 
#filling it with values 1 ,2 based on the values of sex column
df.head()



#x1 = float("nan")
def getNumber(x):
    if x=="Individual":
        return 0
    elif x!=x:
        return 0
    else:
        return 1
df["Reco_Insurance_Type"]=df["Reco_Insurance_Type"].apply(getNumber)



def getNumber(x):
    if x!=x:
        return 0
    else:
        return int(float(x))
df["Holding_Policy_Type"]=df["Holding_Policy_Type"].apply(getNumber)
#We have created a new column called "gender" and 
#filling it with values 1 ,2 based on the values of sex column



df["Reco_Policy_Premium"]=df['Reco_Policy_Premium'].astype(int)



def getNumber(x):
    if x!=x:
        return 0
    else:
        return int(float(x[-1]))
df["Health Indicator"]=df["Health Indicator"].apply(getNumber)
#We have created a new column called "gender" and 
#filling it with values 1 ,2 based on the values of sex column
df.head()



df['diff_between_age']=df['Upper_Age']-df["Lower_Age"]


def getNumber(x):
    if x>14:
        return 1
    else:
        return 0
df["Holding_policy>0"]=df["Holding_Policy_Duration"].apply(getNumber)
#We have created a new column called "gender" and 
#filling it with values 1 ,2 based on the values of sex column
df.head()

df['Region_Code']=df['Region_Code'].astype(int)

def getNumber(x):
    if x!=x:
        return 0
    else:
        return (int(x[1:]))
df["City_Code"]=df["City_Code"].apply(getNumber)

df['Region_Code']=df['Region_Code'].astype(int)


del df['City_code_region']


df.head()
df.to_csv("train.csv",index=False)
```

# Model

As the Data is imbalanced and has a lot of categorical features,I decided to go with Catboost.
as Ensemble models deal with imbalanced dataset well.

I created 3 created Catboost model and used their out of fold prediction to stack them together.

I performed a RandomizedSearchCV on all three Base_classifiers for optimal performance.




Base_classifier1 [CatBoostClassifier(loss_function="Logloss",learning_rate=0.015, 
                             iterations=1400,l2_leaf_reg=1, depth=8, border_count=32,rsm=None,random_state=0,thread_count=4,verbose=1)]



Base_classifier2=[CatBoostClassifier(loss_function="Logloss",learning_rate=0.015, iterations=1300,l2_leaf_reg=7 depth=8, border_count=20,rsm=0.75,random_state=0)]


Base_classifier3=[clf = CatBoostClassifier(loss_function="Logloss",learning_rate=0.041274, iterations=1200,random_state=0)
]



Using output of these three classfiers and appending them to train dataset.(instead of just using outputs of classifier
,appending the with train dataset and predicting with metaclassifier gives better results)

Meta_classifier=[clf = CatBoostClassifier(loss_function="Logloss",learning_rate=0.015, 
                             iterations=1400,l2_leaf_reg=1, depth=8, border_count=32,rsm=None,random_state=0,thread_count=4,verbose=1)
]

```python
df=pd.read_csv('train.csv')
del df['ID']

import matplotlib.pylab as plt
from scipy import interp
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_curve,auc
from sklearn.model_selection import StratifiedKFold,KFold
import matplotlib.patches as patches
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.naive_bayes import BernoulliNB
from catboost import CatBoostClassifier
random_state = np.random.RandomState(0)
#clf = RandomForestClassifier(bootstrap=True, criterion="entropy", max_features=0.7000000000000001, min_samples_leaf=17, min_samples_split=19, n_estimators=150)

cat_features = [0,1,3,6,7,9,10]

clf = CatBoostClassifier(loss_function="Logloss",learning_rate=0.015, 
                             iterations=1400,l2_leaf_reg=1, depth=8, border_count=32,rsm=None,random_state=0,thread_count=4,verbose=1)

y=df['Response']
del df['Response']
x=df

df_test=pd.read_csv('test.csv')
del df_test['ID']

prediction = clf.fit(x,y,cat_features=cat_features).predict_proba(df_test)[:,1]





sub=pd.read_csv('sample_submission_QrCyCoT.csv')
sub["Response"]=prediction
sub.to_csv('1400.csv',index=False)
```

# prediction out-of-fold predictions for  of all base classifiers  and saving them to csv, to input for meta-classifier

```python
oof=pd.DataFrame()

import pandas as pd
df=pd.read_csv('train.csv')



del df['ID']


import matplotlib.pylab as plt
from scipy import interp
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_curve,auc
from sklearn.model_selection import StratifiedKFold,KFold
import matplotlib.patches as patches
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.naive_bayes import BernoulliNB
from catboost import CatBoostClassifier
random_state = np.random.RandomState(0)
#clf = RandomForestClassifier(bootstrap=True, criterion="entropy", max_features=0.7000000000000001, min_samples_leaf=17, min_samples_split=19, n_estimators=150)


clf1 =  CatBoostClassifier(loss_function="Logloss",learning_rate=0.015, 
                             iterations=1400,l2_leaf_reg=1, depth=8, border_count=32,rsm=None,random_state=0,thread_count=4,verbose=1)



clf2 = CatBoostClassifier(
    loss_function="Logloss",learning_rate=0.015, iterations=1300,l2_leaf_reg=7, depth=8, border_count=20,rsm=0.75,random_state=0)




clf3 = CatBoostClassifier(
    loss_function="Logloss",learning_rate=0.041274, iterations=1200,random_state=0)


cat_features = [0,1,3,6,7,9,10]
y=df['Response']
del df['Response']
x=df
ntrain = x.shape[0]



kf = KFold(n_splits = 4, shuffle=False)



def get_oof(clf):
    oof_train = np.zeros((ntrain,))

    for i, (train_index, test_index) in enumerate(kf.split(x)):
        x_tr = x.loc[train_index]
        y_tr = y.loc[train_index]
        x_te = x.loc[test_index]

        clf.fit(x_tr, y_tr,cat_features=cat_features)

        oof_train[test_index] = clf.predict_proba(x_te)[:,1]
        print(oof_train)
    return oof_train.reshape(-1, 1)


clf_1 = get_oof(clf1)
clf_2 = get_oof(clf2)
clf_3 = get_oof(clf3)


#oof.to_csv('oof.csv',index=False)


```

```python
oof['1400']=clf_1[:,0]
oof['1300']=clf_2[:,0]
oof['1200']=clf_3[:,0]

```
